### SEABORN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("titanic-training-data.csv")
df.head()
```

```python
df=df.drop(["PassengerId","Name","Ticket","Cabin"],axis=1)
df.head()
```

## UNIVARIATE

```python
sns.countplot(x="Survived",data=df)
plt.show()
```

```python
sns.countplot(x="Pclass",data=df)
df.head()
```

```python
sns.countplot(x="Embarked",data=df)
df.head()
```

```python
sns.countplot(x="Sex",data=df)
df.head()
```

```python
sns.boxplot(x="Age",data=df)
plt.show() # for numerical attribute
```

```python
sns.boxplot(x="SibSp",data=df)
plt.show()
```

```python
df.describe()
```

```python
sns.boxplot(x="Fare",data=df)
plt.show()
```

```python
sns.boxplot(x="Parch",data=df)
plt.show()
```

```python
plt.hist("Age",data=df)
plt.show()
```

```python
plt.hist("Parch",data=df)
plt.show() 
```

```python
plt.hist("Sex",data=df)
plt.show()
```

```python
# if the column is categorical we use count plot
# if the plot is numerical we use box plot and histogram
# we use box plot when we want to know out lyers
#if we have to study the shape of data then use hist
```
